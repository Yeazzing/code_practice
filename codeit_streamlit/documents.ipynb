## 11. 세션 스테이트로 값 유지하기

세션 스테이트 
* Streamlit한테 값을 맡겨 놓는 기능
* Rerun 하는 동안에도 값이 유지되는 내부의 저장소
* 세션 : 사용자가 웹 페이지에 들어와서 나가기까지의 기간 
* rerun되는 동안은 유지되지만, 웹페이지를 나갔다가 다시 들어오면 초기화

```python
#세션 스테이트 사용하기
import streamlit as st

if "number" not in st.session_state:
    st.session_state["number"] = 1 #처음 실행하는 경우에만 초기화
st.text(st.session_state["number"]) 

clicked = st.button("더하기")
if clicked:  #버튼을 클릭하면 rerun
    st.session_state["number"] += 1 

#세션 스테이트와 rerun() 사용하기
import streamlit as st

if "number" not in st.session_state:
    st.session_state["number"] = 1
st.text(st.session_state["number"])

clicked = st.button("더하기")
if clicked:
    st.session_state["number"] += 1
    st.rerun()  #더해질때마다 바로 보여줄 수있음
```

## 13. 입력 폼

* 사용자가 값을 입력할 때마다 rerun 
* 폼은 인풋들을 모아 놓고 제출 버튼을 눌렀을 때만 rerun
* 반드시 하나의 st.form_submit_button() 배치해야함 

```python
import streamlit as st

data = {}

with st.form("과목 설문 조사"):
  data["date"] = st.date_input("수강 날짜")
  data["name"] = st.text_input("이름")
  data["description"] = st.text_area("전체 소감")
  data["level"] = st.number_input(label="학년", value=1, step=1)
  data["preference"] = st.slider(label="과목 선호도", min_value=1, max_value=10)
  data["difficulty"] = st.radio(
      label="난이도",
      options=["쉬웠다", "보통", "어려웠다"],
  )
  data["is_retaking"] = st.checkbox(label="재수강 여부")
  st.form_submit_button("적용")

st.table(data)
```

## 14. 학점 계산기

```python
def calculate_grade(score):
    if score >= 70:
        return "A"
    if score >= 60:
        return "B"
    if score >= 50:
        return "C"
    return "F"

with st.form("학생 정보"):
  name = st.text_input("이름")
  score = st.number_input("점수")
  st.form_submit_button("제출")

st.table({
    "name": name,
    "score": score,
    "grade": calculate_grade(score),
})
```
